# Library using Investpy
Using the library investpy it is possible to get data of stocks, index, ETFs, commodities, cryptocurrencies and bonds. The advantage to use this library is to get historical information which is very useful to analyzing any asset in financial market. This is a tutorial for you and for me a note which I keep to never forget!

This tutorial follow trading com dados

## Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import investpy
```

## Stocks
It is possible to find stock by using the search mechanism. Any convetional word is suitable. User can search for a word presented in the name of the company

```python
results = investpy.search_quotes(text = 'MDIA', products=['stocks'], countries=['brazil'], n_results = 50)
```

```python
for results in results:
    print(results)
```

```python
mdia = results.retrieve_historical_data(from_date='01/01/2014', to_date = '01/06/2022')
```

```python
#Format OHLC: Open, High, Low, Close
mdia.head()
```

```python
#plotting the MDIA3 performance since 02/01/2014
mdia['Close'].plot()
```

```python
results = investpy.search_quotes(text = 'banco do brasil', products=['stocks'], countries=['brazil'], n_results = 50)
```

## ETFs
The library investpy has a specific function to search ETF. The ETF tries to simulate the performance of a given index.

```python
bova11 = investpy.get_etf_historical_data(etf = 'Ishares Ibovespa', country = 'brazil', from_date = '01/01/2014', to_date = '01/06/2022')
```

```python
bova11['Close'].plot()
```

```python
#gld gold etf
gld = investpy.get_etf_historical_data(etf = 'SPDR Gold Shares', country = 'united states', from_date = '01/01/2014', to_date = '01/06/2022')
```

```python
gld['Close'].plot()
```

```python
# We can search like the very first examples, without using the specific ETF search, however we should change the 'products' to 'etfs' 
results = investpy.search_quotes(text = 'ewz', products=['etfs'], countries=['united states'], n_results = 50)

for results in results[:1]:
    print(results)
```
